```python
"""
https://medium.com/analytics-vidhya/compare-more-than-5-keywords-in-google-trends-search-using-pytrends-3462d6b5ad62
https://pypi.org/project/pytrends/

"""

from pytrends.request import TrendReq
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import time
import datetime
from google.colab import files

pytrend = TrendReq()

normalizationbrand='rapha'
list1=[normalizationbrand,'isadore','assos','la passione','maap']
list2=[normalizationbrand, 'castelli' ,'pas normal studios','cafe du cycliste','le col']
country='SK'
rolling=6


pytrends1=TrendReq()
pytrends2=TrendReq()

# Add sleep time before calling build_payload
#time.sleep(120)
pytrends1.build_payload(list1,timeframe='today 5-y',geo=country)

# Add sleep time before calling build_payload
#time.sleep(120)
pytrends2.build_payload(list2,timeframe='today 5-y',geo=country)

df1=pytrends1.interest_over_time().drop(['isPartial'], axis=1)
df2=pytrends2.interest_over_time().drop(['isPartial'], axis=1)

#normalization
for R in range( df2.shape[0]):
    for C in range( df2.shape[1]-1):
        if C==0:
            if df2.iat[R,C] != 0:
                normalizationFactor = int(df1.iat[R,C])/int(df2.iat[R,C])
            else:
                normalizationFactor = 0
        df2.iat[R,C] = df2.iat[R,C]*normalizationFactor


#merge
df2 = df2.drop([normalizationbrand], axis=1)
result = pd.concat([df1,df2], axis=1)

result= result.rolling(rolling).mean()

fig = px.line(result)

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=5, label="5y", step="year", stepmode="backward"),
            dict(count=10, label="10y", step="year", stepmode="backward"),
            dict(count=20, label="20y", step="year", stepmode="backward"),
        ])
    )
)
fig.show()

fig2 = px.bar(result)

fig2.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=5, label="5y", step="year", stepmode="backward"),
            dict(count=10, label="10y", step="year", stepmode="backward"),
            dict(count=20, label="20y", step="year", stepmode="backward"),
        ])
    )
)
fig2.show()

result.index = result.index.strftime('%Y-%m')
result.info()

current_date = datetime.datetime.now().strftime("%Y-%m")
filename = f"SoS_{country}_{current_date}.csv"

result.to_csv(filename, index=False)
files.download(filename)


```
